## EJERCICIO 5:

Función que recibe lista:

def listar_materiales(materiales): → recibe lista, imprime cada uno numerado

```python
def listar_materiales(materiales):
    materiales_unicos = set(materiales)
    for i, material in enumerate(materiales_unicos):
        print(f"{i+1}. {material}")

materiales = ["cemento", "arena", "grava", "cemento", "agua", "arena"]
listar_materiales(materiales)
```

## EJERCICIO 6:

Función con while + input dentro:

def recolectar_materiales(): → while hasta "fin", append a lista, return la lista

```python
def recolectar_materiales():
    materiales = []
    while True:
        material = input("Ingrese un material (o 'fin' para terminar): ")
        if material.lower().strip() == 'fin':
            break
        materiales.append(material)
    return materiales

materiales_recolectados = recolectar_materiales()
print(f"Materiales recolectados: {materiales_recolectados}")
```

## EJERCICIO 7:

Función que combina todo:

def presupuesto_simple(): → pide material y precio (while), guarda en lista de dicts, calcula total con otra función

```python
def presupuesto_simple():
    total = 0
    print("Presupuesto (escribe 'fin' para terminar)\n")
    
    while True:
        item = input("Material: ").strip()
        if item.lower() == "fin":
            break
        try:
            costo = float(input("Costo S/: "))
            total += costo
            print(f"→ {item}: S/ {costo:.2f}")
        except:
            print("Número inválido, intenta de nuevo")
    
    print(f"\nTotal final: S/ {total:.2f}")
```

def clasificar_viga(longitud): → "corta" <3m, "media" 3–6m, "larga" >6m

```python
def clasificar_viga(longitud):
    if longitud < 3:
        return "Corta"
    elif 3 <= longitud <= 6:
        return "Media"
    else:
        return "Larga"

viga1 = clasificar_viga(4.5)
viga2 = clasificar_viga(7.2)
print(f"La viga1 es: {viga1}")
print(f"La viga2 es: {viga2}")
```

## EJERCICIO 10:

def maximo_precio(materiales): → recibe lista de dicts, devuelve el material más caro

```python
def maximo_precio(materiales):
    """
    Recibe una lista de diccionarios con 'nombre' y 'costo' (o 'precio')
    Devuelve el nombre del material con el precio más alto
    """
    if not materiales:
        return "No hay materiales"
    
    material_mas_caro = max(materiales, key=lambda x: x["costo"])
    return material_mas_caro["nombre"]

# Ejemplo de uso
presupuesto = [
    {"nombre": "cemento",        "costo": 450},
    {"nombre": "fierro",         "costo": 1200},
    {"nombre": "arena fina",     "costo": 380},
    {"nombre": "bloques",        "costo": 680},
    {"nombre": "alambre",        "costo": 120},
]

print("Material más caro:", maximo_precio(presupuesto))
```

## EJERCICIO 11:

Ejercicios más integrados

def calcular_volumenes(elementos): → recibe lista de dicts [{'tipo': 'viga', 'largo':5, 'ancho':0.3, 'alto':0.5}, ...] → calcula volumen por elemento y total

```python
def calcular_volumenes(elementos):
    """
    Recibe una lista de diccionarios con 'tipo', 'largo', 'ancho', 'alto'
    Devuelve el volumen total de todos los elementos
    """
    if not elementos:
        return "No hay elementos"

    volumen_elemento = lambda e: e['largo'] * e['ancho'] * e['alto']
    return sum(volumen_elemento(e) for e in elementos)

elemento_concreto = [
    {"tipo": "viga", "largo": 5, "ancho": 0.3, "alto": 0.5},
    {"tipo": "columna", "largo": 3, "ancho": 0.4, "alto": 0.4},
    {"tipo": "losa", "largo": 6, "ancho": 4, "alto": 0.15},
]

print(f"Volumen total de elementos: {calcular_volumenes(elemento_concreto)} m³")
```
